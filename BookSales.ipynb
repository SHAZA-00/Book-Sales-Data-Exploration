```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/sample_data/Books_Data_Clean.csv")
df.head()
```

```python
df.describe()
```

```python
df = df[df["Publishing Year"] > 1900]
```

```python
df.isna().sum()
```

```python
df.dropna(subset= "Book Name", inplace= True)
```

```python
df.duplicated().sum()
```

```python
df.nunique()
```

```python
plt.hist(df['Publishing Year'])
plt.xlabel('Publishing Year')
plt.ylabel('FRQ')
plt.title('Distribution of publishing year')
plt.show()
```

```python
df['genre'].value_counts().plot(kind='bar')
plt.xlabel('Genre')
plt.ylabel('Numbers')
plt.title('Numbers of Genre Count')
```

```python
df.groupby('Author')['Book_average_rating'].mean().sort_values(ascending=False)
```

```python
sns.boxplot(x='genre', y='Book_ratings_count', data= df)
plt.xlabel('Genre')
plt.ylabel('Book Rating Counts')
plt.title('Box Plot for Book Rating Count')
plt.show()
```

```python
plt.scatter(df['sale price'], df['units sold'])
plt.xlabel('Sale Price')
plt.ylabel('Units Sold')
plt.title('Sale Price vs Units Sold')
plt.show()
```

```python
langugae_count = df['language_code'].value_counts()
plt.pie(langugae_count, labels=langugae_count.index, startangle=90, autopct="%1.1f%%")
plt.title("Language Distribution of Books")
plt.show()
```

```python
df.columns
```

```python
df.groupby('language_code').size().sort_values(ascending=False)
```

```python
plt.scatter(df['Book_average_rating'], df['Book_ratings_count'])
plt.xlabel('Book Average Rating')
plt.ylabel('Book Ratings Count')
plt.title('Scatter Plot for Book average Rating vs. Book rating count')
plt.show()
```

```python
total_gross_sales = df.groupby('Author')['gross sales'].sum()
```

```python
total_gross_sales.sort_values(ascending=False).head().plot(kind='bar')
plt.xlabel('Author')
plt.ylabel('Total Sales')
plt.title('Total Sales for each Author')
```

```python
unit_solt_each_publishing_year = df.groupby('Publishing Year')['units sold'].sum()
unit_solt_each_publishing_year.plot(kind='line', marker='o')
plt.xlabel('Publishing Year')
plt.ylabel('Units Sold')
plt.title('Total Units Sold each Publishing Year')
```
